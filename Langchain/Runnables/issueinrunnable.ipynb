```python
import random
class Naklillm:
    def __init__(self):
        print('llm created')
    def predict(self,prompt):
        response_list=['delhi is capital of india','ipl is indian premier league','ai is ml']
        return {'response':random.choice(response_list)}
```

```python
llm=Naklillm()
llm.predict("what is capital of indiaa")
```

```python
class NaklipromptTemplate:
    def __init__(self,template,input_variables):
        self.template=template
        self.input_variables=input_variables
    def format(self,input_dict):
        return self.template.format(**input_dict)
```

```python
template=NaklipromptTemplate(
    template='write a poem on {topic}',
    input_variables=['topic']
)
```

```python
prompt=template.format({'topic':'india'})
```

```python
llm.predict(prompt)
```

```python
class naklillmchain:
    def __init__(self,llm,template):
        self.llm=llm
        self.template=template
    def run(self,input_dict):
        final_prompt=self.template.format(input_dict)
        result=self.llm.predict(final_prompt)
        return result['response']
        
```

```python
chain=naklillmchain(llm,template)
chain.run({'topic':'ai'})
```
